```python
import numpy as np
import matplotlib.pyplot as plt
import random
np.random.seed(10)
```

```python
X_ = np.sort(np.random.uniform(low = 0.0, high = 1.0, size = 500))
Y_ = np.sin(2 * np.pi * X_)
noise_ = + np.random.normal(  scale = 0.3, size = Y_.shape)
YY_ = Y_ + noise_

# y = sin(2piX) + GN(gausian noise)
```

```python
random.seed(10)
ind_ = random.sample(range(500), 500)
ind_train = ind_[:10]
X_sub_train = X_[ind_train]
YY_sub_train = YY_[ind_train]
```

```python
# Linear regression
# M = 1
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
L_1 = LinearRegression()
L_1.fit(X_sub_train[:,np.newaxis], YY_sub_train)
predictions = L_1.predict(X_[:,np.newaxis])
```

```python
plt.plot(X_,YY_, color = 'green')
plt.plot(X_,Y_, color = 'red')
plt.plot(X_,predictions)
plt.show()
```
