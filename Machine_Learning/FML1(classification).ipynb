```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

### Creating a dataframe ###
df = pd.DataFrame([
    [1.0,2.0,3.0,4.0],
    [12.2,2.3,4.3,2.3],
    [np.NaN,3.4,2.4,np.NaN]
])

df.columns = ['A','B','C','D']
print(df)
print(df.isnull().sum())

df_1 = df.dropna(axis = 0) ## Drop rows
print(df_1)
df_2 = df.dropna(axis = 1) ## Drop columns
print(df_2)

##Only drop rows where all the columns are NaN
df_3 = df.dropna(how = 'all')

##Only drop rows where NaN appears in specific column
df.dropna(subset = ['A'])
```

```python
df = pd.DataFrame([
    ['green','M',10.1,'class1'],
    ['red','L',13.5,'class2'],
    ['blue','XL',15.3,'class1']
])

df.columns = ['color','size','price','classlabel']
print(df)
```

```python
#Nominal vs ordinal features
#Mapping Ordinal features
size_mapping = {
    'XL':3,
    'L':2,
    'M':1,
}

df['size'] = df['size'].map(size_mapping)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
y = class_le.fit_transform(df['classlabel'].values)

#print(y)

df['classlabel'] = y
print(df)
# inverse transform
R_ = class_le.inverse_transform(y)
print(R_)
```

```python
#onehotEncoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output= False).set_output(transform = 'pandas')
F_ = ohe.fit_transform(df[['color']])
print('FF__')
##print(F_)

#YY_ = pd.concat([df,F_],axis = 1)
YY_ = pd.concat([df,F_],axis = 1).drop('color',axis = 1)
print(YY_)
```
