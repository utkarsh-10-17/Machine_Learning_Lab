```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.drop('Id',axis=1,inplace=True)
columns = df.columns.tolist()
Classes = (df['Species'].unique()).tolist()
#result = (df.loc[df['Species']]) == Classes[0],columns[0]
#print(result)
print(Classes)
```

```python
df.head()
```

```python
def get_pdf_par(attribute, class_):
    result = (df.loc[df['Species'] == class_,attribute])
    mean = np.mean(result)
    var = np.std(result)**2
    return mean, var
```

```python
def get_likelihood(value_,mean,var):
    return (1/(np.sqrt(2*np.pi*var)))*np.exp(-(value_-mean)**2/(2*var))
```

```python
T = {'SepalLengthCm':5 , 'SepalWidthCm':2,'PetalLengthCm':3.5,'PetalWidthCm':1}
print(T.keys())
S = []
for j in Classes:
  L = 0;
  for i in T.keys():
    mean,var = get_pdf_par(i,j)
    Likelihood = np.log(get_likelihood(T[i],mean,var))
    L = L + Likelihood
  S.append(L)
  ind_1 = np.where(S == np.max(S))
  predicted_class = Classes[int(ind_1[0])]
  print(predicted_class)

print(S)
```
